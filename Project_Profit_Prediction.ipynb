```python
x = df.drop(['Profit'],axis = 1) # independent variable
```

```python
x
```

```python
y = df['Profit'] # dependent 
```

```python
y
```

```python
# Graph shown
%matplotlib inline
plt.xlabel('Transport')
plt.ylabel('Profit')
plt.scatter(df.Transport, df.Profit,color='red',marker='+')
```

```python
# one hot encoding
#coverting coloums into catagorical columns
city = pd.get_dummies(x['Area'],drop_first=True)
```

```python
city
```

```python
# Drop the area column
x = x.drop("Area",axis = 1)
```

```python
x
```

```python
#concatation
x = pd.concat([x,city],axis = 1)
```

```python
x
```

```python
#splitting the dataset into the trainning set and test set
#import library

from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state=0)
```

```python
# fitting linear 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
ytest
```

```python
# predict the resulting value

pred = reg.predict(xtest)
```

```python
pred
```

```python
# Acuracy

reg.score(xtest,ytest)
```

```python
# Alternative way -> accuracy
# R - squared value

from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest, pred)
```

```python
score
```

```python
%matplotlib inline
plt.scatter(xtest, ytest, color = 'blue')
plt.scatter(xtest, pred, color = 'red')
y_line = reg.prediction(x)
plt.plot(x, y_line,'black')
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.title('Actual Data vs Predicted Data')
```
